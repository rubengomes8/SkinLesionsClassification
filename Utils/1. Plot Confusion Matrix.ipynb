# Step 1

Run the following cell

```python
no_classes = 7
path = '/home/ruben/Desktop/isic_2018/results.txt'

with open(path, "r") as f:
    data = f.readlines()
    for i in range(0, no_classes):
        data[i] = data[i][:-1].split(" ")


data = data[0:7]
s = '['
r = 0
for row in data:
    col = 0
    s = s + '['
    for c in row:
        if col < 6:
            s = s + c + ','
        elif col == 6:
            if r < 6:
                s = s + c + '],\n'
            elif r == 6:
                s = s + c + ']'
        col += 1
    r += 1

print(s+']')
```

# Step 2: Plot Confusion Matrix

* Now, paste the content in the initialization of conf_mat

```python
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
import numpy as np


multiclass = np.array([[51,7,6,6,14,7,3],
[137,422,8,5,26,6,17],
[1,1,22,5,3,0,8],
[0,0,12,11,1,0,0],
[16,3,9,4,52,3,7],
[0,0,1,0,2,1,3],
[0,0,2,0,0,0,10]])

class_names = ['MEL', 'NV', 'BKL', 'DF', 'VASC', 'BCC', 'AKIEC']

fig, ax = plot_confusion_matrix(conf_mat=multiclass,
                                colorbar=True,
                                show_absolute=False,
                                show_normed=True,
                                class_names=class_names)
plt.show()
```
